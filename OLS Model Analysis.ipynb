```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
from scipy import stats
from statsmodels.graphics.regressionplots import plot_regress_exog, plot_fit, plot_leverage_resid2, influence_plot
```

# Set 1:

```python
set_1 = pd.read_excel(r'C:\Users\Admin\Desktop\linear_regresion_data.xlsx', sheet_name = 'Set 1')
pd.plotting.scatter_matrix(set_1)

ols_1 = sm.ols('y ~ x', data = set_1)
ols_1 = ols_1.fit()
ols_1.summary()
```

The scatter plots visually show a linear relationship. Frpm the OLS model, we see that the intercept is ~0.24 and the slope is ~4.98; we also get an R squared of 0.979 implying that 97.9% the data is close to the fitted regression line (as seen below). 

```python
plt.scatter(set_1['x'], set_1['y'], label = 'Original data')
plt.title('Regression line along Set 1 data')
intercept, slope = ols_1.params
set_1['predicted'] = intercept + slope * set_1.x
plt.plot(set_1.x, set_1.predicted , 'r-', label='Fitted Line')
plt.legend()
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```

We observe that the residuals are approximately centered around 0 and the model is visually linear around the regression line, as per the graph above, with no outliers.

```python
set_1['Residuals'] = set_1['predicted'] - set_1['y']
residuals_hist_1 = plt.hist(set_1.Residuals)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Residuals of Set 1')
```

We can varify the variance of the points below by graphing the residuals.We can visually check off that 
there is constant variance of the response variable 𝑦 regardless of the value of the independent variable 𝑥.

```python
plt.scatter(set_1['x'], set_1['Residuals'])
plt.xlabel('x')
plt.ylabel('Residuals')
plt.title('Residuals of Set 1 against X values')
```

# Set 4:

```python
set_4 = pd.read_excel(r'C:\Users\Admin\Desktop\linear_regresion_data.xlsx', sheet_name = 'Set 4')
pd.plotting.scatter_matrix(set_4)

ols_4 = sm.ols('Y ~ x', data = set_4)
ols_4 = ols_4.fit()
ols_4.summary()
```

The scatter plots show a nonlinear relationship between the x and Y values. We get extreme high values for the intercept and slope as well as a R squared value of 0.38 implying that only 38% of the data is close to the fitted regression line (seen below).

We also note 2 obvious outliers.

```python
plt.scatter(set_4['x'], set_4['Y'], label = 'Original data')
plt.title('Regression line along Set 4 data')
intercept, slope = ols_4.params
set_4['predicted'] = intercept + slope * set_4.x
plt.plot(set_4.x, set_4.predicted , 'r-', label='Fitted Line')
plt.legend()
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```

We observe that the residuals are mostly centered around 0 but does not depict a normal distribution as it is cut off at the far right. 

```python
set_4['Residuals'] = set_4['predicted'] - set_4['Y']
residuals_hist_4 = plt.hist(set_4.Residuals)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Residuals of Set 4')
```

We see that the residual points are very linear implying that this data violates the constant variance assumption of OLS regression/

```python
plt.scatter(set_4['x'], set_4['Residuals'])
plt.xlabel('x')
plt.ylabel('Residuals')
plt.title('Residuals of Set 4 against X values')
```

## SET 4 Without Outliers:

Getting rid of the two outliers will vastly change the outcome of the OLS model.

```python
z_4 = np.abs(stats.zscore(set_4))
set_4_no_out = set_4[(z_4 < 3).all(axis=1)]
```

In the lines of code above, we are removing any data points that have a z score greater than 3 because we know that almost all of the data (99.7%) should be within 3 standard deviations from the mean, anything more is considered an outlier.

```python
ols_4_out = sm.ols('Y ~ x', data = set_4_no_out)
ols_4_out = ols_4_out.fit()
ols_4_out.summary()
```

We see that the intercept went from -7.535e+05 to -5.558e+05, the slope from 2.707e+05 to 2.106e+05, and R squared went from 38% to 47.6%. Since removing the outlier changed the slope by quite a bit, we deem the outlier to be an influential point but since it is following the general trend of the graph and does not have extreme x values, it does not have high leverage.
